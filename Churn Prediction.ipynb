```python
Data Cleaning
```

```python
Creating one copy of data for Manipulation and processing
```

```python
telco_data = telco_base_data.copy()
```

```python
Total charger should be numeric value. Let convet it to numeric data type
```

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors = "coerce")
telco_data.isnull().sum()
```

```python
# Let check these 11 missing values
telco_data.loc[telco_data["TotalCharges"].isnull() == True]
```

```python
Since the % of the records compared to total dataset is very low i.e 0.15%, it is safe to ignore them for further processing.
```

```python
# Removing missing values
telco_data.dropna( how = "any", inplace = True)
```

```python
 Divide customers into bins based on tenure eg/ for tenure < 12 months: assign a tenure group if 1-12 for tenure between 1 to 2
Years, tenure group of 13-24; and so on...
```

```python
# Get the max tenure 
print(telco_data["tenure"].max())
```

```python
# Group the tenure in bins for 12 months 
labels = ["{0} - {1}".format(i, i+11) for i in range(1,72,12)]

telco_data["tenure_group"] = pd.cut(telco_data.tenure, range(1,80,12), right = False, labels = labels)
```

```python
telco_data["tenure_group"].value_counts()
```

```python
# Removing columns that are not required
telco_data.drop(columns = ["customerID", "tenure"], axis = 1, inplace = True)
telco_data.head()
```

```python
Exploratory Data Analysis
```

```python
for i, predictor in enumerate(telco_data.drop(columns = ["Churn", "TotalCharges", "MonthlyCharges"])):
    plt.figure(i)
    sns.countplot(data = telco_data, x = predictor, hue = "Churn")
```

```python
Convert the target variable "Churn" in a binary numeric variable i.e. Yes = 1, No = 0
```

```python
telco_data["Churn"] = np.where(telco_data.Churn == "Yes", 1, 0)
telco_data.head()
```

```python
Now converting categorical variables into dummy variables
```

```python
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style("whitegrid")
    sns.set_context("talk")

    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams["axes.titlesize"] = 22
    plt.rcParams["axes.titlepad"] = 30

    if hue:
        temp = df[hue]
        width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    else:
        width = len(df[col].unique()) + 7

    fig, ax = plt.subplots()
    fig.set_size_inches(width, 8)

    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)

    sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index,
        hue=hue,
        palette='bright',
        ax=ax
    )

    plt.show()
```

```python
uniplot(new_df1_target1, col = "Partner", title= "Distribution  of Gender for Churned Customers", hue = "gender")
```

```python
uniplot(new_df1_target0, col = "Partner", title= "Distribution  of Gender for Churned Customers", hue = "gender")
```
